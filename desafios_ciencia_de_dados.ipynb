```python
import pandas as pd
import matplotlib.pyplot as plt

url_dados = 'https://github.com/alura-cursos/imersaodados3/blob/main/dados/dados_experimentos.zip?raw=true'

dados = pd.read_csv(url_dados, compression = 'zip')
dados
```

```python
#Dados inferidos com uma percepção inicial da tabela

print('Quantidade de id: ', dados['id'].nunique())
print('Tratamento: ', dados['tratamento'].unique())
print('Tempo em horas: ', dados['tempo'].unique())
print('Doses: ', dados['dose'].unique())
print('Número de drogas que foram investigadas: ', dados['droga'].nunique())
```

**Desafio 03: Proporção das classes "tratamento"**

```python
dados = dados['tratamento'].value_counts().plot.pie(colors = ['#7B68EE', '#FF7F50'], autopct='%1.1f%%')

plt.title('Proporção de tratamento')
plt.legend()
```

**Desafio 05: Procurar na documentação o método query (pandas).**

**Outros gráficos:**

```python
valor_por_tempo = dados.tempo.value_counts()
cores = ['#98FB98', '#AFEEEE', '#DB7093']

plt.title('Tempo de exposição ao tratamento (em horas)')
plt.xlabel('Tempo (em horas)')
plt.ylabel('Quantidade de tratamentos')

ax = valor_por_tempo.sort_index().plot.bar(color = cores)
```
